```python
# Importing Important Libraries
import pandas as pd
import numpy as npy
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Loading the dataset
infile = 'bank-additional-full.csv'
data = pd.read_csv(infile, sep=';', low_memory=False)
```

```python
# Plot 1
data_new = data[data['y']=='yes']
data_new.replace(['yes', 'no', 'unknown', 'single', 'married', 'divorced'], 
                     ['Yes', 'No', 'Unknown', 'Single', 'Married', 'Divorced'], inplace=True)
with sns.plotting_context("notebook",font_scale=1.2):
    g = sns.catplot(x="y", y="previous", hue="month", col="marital", row="loan", kind="bar", data=data_new, capsize=.01)
    g.set_axis_labels("Customers Subscribing to Term Deposit", "No. of Times Customer Previously Contacted")
    g.set_titles("Loan: {row_name} | Marriage Status: {col_name}")
plt.show()
plt.savefig('Plot1.png', dpi=300)
```

```python
# Plot 2
with sns.plotting_context("notebook",font_scale=1.2):
    g = sns.catplot(y="y", hue="education", kind="count", col="housing", row="default",
                palette="pastel", edgecolor=".6",
                data=data_new)
    g.set_axis_labels("Count", "Customers Subscribing to Term Deposit")
    g.set_yticklabels("")
    g.set_titles("Credit Default: {row_name} | Housing: {col_name}")
plt.show()
plt.savefig('Plot2.png', dpi=300)
```
